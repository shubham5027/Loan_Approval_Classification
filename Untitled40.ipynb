```python
from sklearn.metrics import precision_score,accuracy_score

def trainer(model,X_train,X_test,y_train,y_test):
  model.fit(X_train,y_train)
  y_pred=model.predict(X_test)
  precision=precision_score(y_test,y_pred)
  accuracy=accuracy_score(y_test,y_pred)
  return precision,accuracy
```

```python
print(trainer(lr,X_train,X_test,y_train,y_test))
print(trainer(dtc,X_train,X_test,y_train,y_test))
print(trainer(rfc,X_train,X_test,y_train,y_test))
print(trainer(knn,X_train,X_test,y_train,y_test))
print(trainer(svc,X_train,X_test,y_train,y_test))
print(trainer(nb,X_train,X_test,y_train,y_test))
print(trainer(xg,X_train,X_test,y_train,y_test))
```
